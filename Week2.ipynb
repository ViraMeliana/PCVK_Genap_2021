```
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #imagedisplay
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
```

```
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
       "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
       "https://placekitten.com/800/571"]

for url in urls:
  image = io.imread(url)                                # Read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)       # resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)        # convert color to RGB
  final_frame = cv.hconcat((image, image_2))            # concatenate image
  cv2_imshow(final_frame)                               # show image
  print('\n')
```

```
length = image_2.shape[0]
width = image_2.shape[1]
print("Image Resolution: Length x Width = ",length, " x ",width)
cv2_imshow(image_2)
```

```
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Create horizontal line in the center of image
for y in range (width):
  image_3[int((length)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

  > Further, the byte order in 32 bit integers is so that the first 16 bit word will be the LSB bits. Thus looking at the hex representation of your two first words.
```
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
img3 = img
top = plt.get_cmap('Reds', 50) # r means reversed version
bottom = plt.get_cmap('Greens', 50)# combine it all
newcolors = np.vstack((top(np.linspace(0, 1, 128)),
                       bottom(np.linspace(0, 1, 128))))
red_greens = ListedColormap(newcolors, name='red_greens')
plt.imshow(img3, red_greens)
```

```
img4 = img
top2 = plt.get_cmap('Greens', 250) # r means reversed version
bottom2 = plt.get_cmap('Blues', 250)# combine it all
newcolors = np.vstack((top2(np.linspace(0, 1, 250)),
                       bottom2(np.linspace(0, 1, 250))))
green_blue = ListedColormap(newcolors, name='green_blue')
plt.imshow(img4, green_blue)
```
